```python
import scanpy as sc
import pandas as pd
import seaborn as sns
import numpy as np
import anndata as ad
from matplotlib import pyplot as plt
```

```python
# please specify which dataset you want to process:
# available options: PBMC, WBC, PFC
dataset = "WBC"

# please specifiy path for reading input files
PATH = dataset.lower()

# Please specify which tool you used
# available options: cellbender, cellclear, scar, soupx, decontx, uncorrected
tool = "uncorrected"
```

```python
adata = sc.read_h5ad(f"{PATH}/{dataset}_{tool}_processed.h5ad")
adata
```

```python
adata = adata[adata.obs["predicted_doublet"] == False]
```

### Confidence scores

```python
if dataset == "PFC":
    df = adata.obs[["celltypist_conf_score"]].copy()
    df["Tool"] = tool
    df.to_csv(f"../{PATH}/conf_scores_{tool}.csv", index=False)

elif dataset == "WBC" or dataset == "PBMC":
    df1 = adata.obs[["celltypist_conf_score_fine"]].copy()
    df1["Tool"] = tool
    df1.to_csv(f"../{PATH}/conf_scores_fine_{tool}.csv", index=False)

    df2 = adata.obs[["celltypist_conf_score_coarse"]].copy()
    df2["Tool"] = tool
    df2.to_csv(f"../{PATH}/conf_scores_coarse_{tool}.csv", index=False)
```

```python
# after extracting all confidence score information for each Tool and without correction (7 times total) combine all files into one
from glob import glob

if dataset == "PFC":
    files = glob(f"../{PATH}/conf_scores*.csv")
    dfs = [pd.read_csv(f) for f in files]
    combined = pd.concat(dfs, ignore_index=True)
    combined.to_csv(f"../{PATH}/{dataset.lower()}_confidence_scores.csv")

elif dataset == "WBC" or dataset == "PBMC":
    files = glob(f"../{PATH}/conf_scores_coarse*.csv")
    dfs = [pd.read_csv(f) for f in files]
    combined = pd.concat(dfs, ignore_index=True)
    combined.to_csv(f"../{PATH}/{dataset.lower()}_confidence_scores_coarse.csv")

    files1 = glob(f"../{PATH}/conf_scores_fine*.csv")
    dfs1 = [pd.read_csv(f) for f in files1]
    combined1 = pd.concat(dfs1, ignore_index=True)
    combined1.to_csv(f"../{PATH}/{dataset.lower()}_confidence_scores_fine.csv")
```

```python
# map fine cell types to broader categories for marker gene analysis
celltype_categories = {
    "Astrocytes": {
        "marker_genes": ["GFAP", "AQP4", "SLC1A2", "ALDH1L1", "FABP7"], # GFAP (highly)
        "contains": ["Astro"]
    },
    "Oligodendrocytes": {
        "marker_genes": ["MOG", "MBP", "PDGFRA", "CSPG4", "OLIG1", "OLIG2", "OPALIN"], # MOG (highly), OLIG2 (moderate), "PDGFRA"
        "contains": ["Oligo"]
    },
    "OPCs": {
        "marker_genes": ["MOG", "MBP", "PDGFRA", "CSPG4", "OLIG1", "OLIG2", "OPALIN"], # MOG (highly), OLIG2 (moderate), "PDGFRA"
        "contains": ["OPC"]
    },
    "Microglia": {
        "marker_genes": ["P2RY12", "CX3CR1", "TMEM119", "APBB1IP", "ITGAM"], # P2RY12
        "contains": ["Micro"]
    },
    "Endothelial cells": {
        "marker_genes": ["CLDN5", "SLC2A1", "SLC7A5", "PECAM1"], # CLDN5 (highly) (rare celltype)
        "contains": ["Endo"]
    },
    "Excitatory neurons": {
        "marker_genes": ["SLC17A7", "SATB2", "FEZF2", "BCL11B", "CUX2", "RORB"], # SLC17A7
        "contains": ["L2", "L3", "L4", "L5", "L6"]
    },
    "Inhibitory Interneurons": {
        "marker_genes": ["GAD1", "GAD2", "SST", "VIP", "PVALB", "LAMP5", "RELN"], # GAD1
        "contains": ["InN"]
    },
    "Pericytes": { 
        "marker_genes": ["PDGFRB", "ABCC9"], # ABCC9
        "contains": ["PC"]
    },
    "COP": { # committed OPCs
        "marker_genes": ["GPR17", "SOX4", "PDGFRA", "CSPG4"],  # GPR17
        "contains": ["COP"]
    },
    "Macrophages": {
        "marker_genes": ["COL1A2", "F13A1"],
        "contains": ["Macro"]
    },
    "VLMCs": { #leptomeningial cells/fibroblast like
        "marker_genes": ["COL1A1"],
        "contains": ["VLMC"]
    },
#    "T cells": {
#        "marker_genes": ["CD3", "CD274"],
#        "contains": ["T"]
#    }
}


def map_celltypes_to_categories(adata, source_column="celltypist_cell_label", target_column="celltype_category"):
    """
    Orders cell type labels in adata.obs[source_column] in broader categories and saves them in adata.obs[target_column]
    """
    def assign_category(celltype_label):
        for category, info in celltype_categories.items():
            if any(keyword in celltype_label for keyword in info["contains"]):
                return category
        return "Not mapped"

    adata.obs[target_column] = adata.obs[source_column].apply(assign_category)
```
